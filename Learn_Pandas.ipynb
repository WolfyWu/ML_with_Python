```python
def get_column_nums(x):
    new = list(x.columns)
    new1 = pd.DataFrame()
    for index, value in enumerate(new):
        temp = pd.DataFrame({'col_name':[value],'col_num':[index+1]})
        new1 = pd.concat([new1,temp])

    return new1
```
