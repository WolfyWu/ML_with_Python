```python
### Weights of players

fifa['Weight']
```

```python
fifa['Weight_kg'] = fifa['Weight'].str.slice(stop=3).astype('float') * 0.453592
```

```python
fifa['Weight_kg']
```

```python
light_players = fifa.loc[fifa['Weight_kg'] < 70].count()[0]

light_medium_players = fifa.loc[(fifa['Weight_kg'] >= 70) & 
                                (fifa['Weight_kg'] < 90)].count()[0]

heavy_players = fifa.loc[fifa['Weight_kg'] >= 90].count()[0]
```

```python
print(light_players, light_medium_players, heavy_players)
```

```python
weights = [light_players, light_medium_players, heavy_players]
```

```python
print(plt.style.available)
```

```python
# Lets plot

labels = ['<70', '70 - 90', '>90']

plt.figure(figsize=(10, 10))

plt.style.use('ggplot')

plt.pie(weights, labels=labels, autopct='%.2f %%')

plt.title('Weights Distribution')

plt.show()
```

```python
# Move around data labels

labels = ['<70', '70 - 90', '>90']

plt.figure(figsize=(10, 10))

plt.style.use('ggplot')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.9)

plt.title('Weights Distribution')

plt.show()
```

```python
# Explode the chart

labels = ['<70', '70 - 90', '>90']

plt.figure(figsize=(10, 10))

plt.style.use('ggplot')

explode = [0,0,.4]

plt.pie(weights, labels=labels, autopct='%.2f %%', 
        pctdistance=0.9, explode=explode)

plt.title('Weights Distribution')

plt.show()
```

```python
manu = fifa.loc[fifa['Club']=="Manchester United"]["Overall"]

plt.figure(figsize=(10,10))

plt.style.use('default')

labels = ['FC Barcelona', 'Real Madrid', "Manchester United"]

plt.boxplot([barcelona, madrid, manu], labels=labels)

plt.show()
```

```python
### Adding style elements
plt.figure(figsize=(10,10))

plt.style.use('default')

labels = ['FC Barcelona', 'Real Madrid', "Manchester United"]

boxes = plt.boxplot([barcelona, madrid, manu], labels=labels)

for box in boxes['boxes']:
    box.set(color="#ff9900")

plt.show()
```

```python
### Changing width
plt.figure(figsize=(10,10))

plt.style.use('default')

labels = ['FC Barcelona', 'Real Madrid', "Manchester United"]

boxes = plt.boxplot([barcelona, madrid, manu], labels=labels)

for box in boxes['boxes']:
    box.set(color="#ff9900", linewidth=2)

plt.show()
```

```python
### Fill boxes
plt.figure(figsize=(10,10))

plt.style.use('default')

labels = ['FC Barcelona', 'Real Madrid', "Manchester United"]

boxes = plt.boxplot([barcelona, madrid, manu], labels=labels, patch_artist=True)

for box in boxes['boxes']:
    box.set(color="#ff9900", linewidth=2)
    box.set(facecolor="#ff9900")
    
plt.show()
```

```python
### Change median line
plt.figure(figsize=(10,10))

plt.style.use('default')

labels = ['FC Barcelona', 'Real Madrid', "Manchester United"]

boxes = plt.boxplot([barcelona, madrid, manu], labels=labels, patch_artist=True, medianprops={'linewidth':4})

for box in boxes['boxes']:
    box.set(color="#ff9900", linewidth=2)
    box.set(facecolor="#F9F608")
    
plt.show()
```

```python
import random, numpy as np
age = list(range(1,101))
weight_male = list(np.random.uniform(20,80,100))
weight_female = list(np.random.uniform(15,75,100))
```

```python
weight_data = pd.DataFrame({'Age':age, 
                            'Weight_male':weight_male,
                            'Weight_female':weight_female})
```

```python
weight_data
```

```python
plt.figure(figsize=(20,20))
plt.bar(weight_data.Age,weight_data.Weight_male)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,15))
index = np.arange(100)
bar_width = 0.35

plt.bar(index, weight_data.Weight_male, bar_width,
                  color='gold',label='Weight_male')
plt.bar(index+bar_width, weight_data.Weight_female, bar_width,
                  color='#CD853F',label='Weight_female')

plt.xlabel('Age')
plt.ylabel('Weights')
plt.title('Weights by Age & Gender')
plt.xticks((index + bar_width)[::9], weight_data.Age[::9])
plt.legend()

plt.tight_layout()
plt.show()
```
